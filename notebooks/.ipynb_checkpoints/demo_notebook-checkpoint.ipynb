## Step 1: Upload Images and Create Dataframes

These functions upload multiple experiments of microscopy data from the same crystal conducted at varying temperatures and create a list of dataframes corresponding to the collection of experiments.

1. **load_images_from_folder:** Upload images from raw files and convert into scikit-image arrays
2. **calculate_channel_intensities:** Extract mean intensity of desired channels from scikit-image arrays and convert into pandas dataframe
3. **add_temp_data:** Interpolate temperature data from .csv file to assign a temperature to each timepoint in the dataframe

**Requirements:**
* Each experiment's images should be stored in their own folder with the same naming convention and the last 3 digits of the folder name corresponding to the experimental temperature
* The individual experiments' folders should be stored in a single enclosing folder (here, *'batch_images'*)
* Default image extension is .png but this can be changed in the function argument
* The timestamp for each image is extracted using naming conventions from Zeiss ZEN blue edition, which consist of a lowercase *t* and then a number (does not need leading 0) that is multiplied by the defined time interval below
* Temperature data for each experiment should be recorded in a .csv file with the same naming convention and the last 3 digits of the folder name corresponding to the experimental temperature
* For each temperature dataset, there should be two columns in the .csv with headings *Time (s)* and *Temperature (°C)* (temperature should be recorded in/converted to °C)
* **Important: the data files need to be in the same folder as this notebook!**

```python
temps_list = ['075', '090', '105', '120', '135'] # enter temperatures using conventions of file names
folder_root = 'demo_data/batch_images/TOPA-1-' # enter filepath for batch image folders
temp_path = 'demo_data/temp_data/TOPA-1-' # enter filepath for temperature .csv data

time = 0.484 # enter time interval for each microscopy image (default for 'crystals_spectral' is 0.484 s)

# ---- Do not edit code below this line --- #

# Call the functions from crystal_tools for use in this code
from crystal_tools.image_loading import load_images_from_folder
from crystal_tools.data_merging import calculate_channel_intensities, add_temp_data

# Create an empty list to store resulting dataframes in
df_list = []

for temp in temps_list:
    # Load images from folder
    images, files = load_images_from_folder(f'{folder_root}{temp}', file_extension='.png')
    df = calculate_channel_intensities(images, channels={'red':0, 'green': 1, 'blue': 2}, time_interval=time)

    # Add temperature data
    temp_filename = f'{temp_path}{temp}.csv'
    add_temp_data(temp_filename, df, time)

    # Add experiment to df_list
    df_list.append(df)
```

## Step 2: Reaction Analysis

This function calculates two useful parameters for further thermodynamics analysis, α (conversion ratio) and Q (reaction quotient) using the function **calculate_rxn_progress**. The function **load_init_image** is necessary for uploading an image at room temperature to obtain baseline intensities of the initial and final channels for normalization.

**Equations:**

* $α(t) = 1 - \dfrac{I_{init}(t)}{I_{final}(t)} * \dfrac{I_{final_0}}{I_{init_0}}$

where $I_{init}(t)$ is the intensity of the initial channel as a function of time, $I_{final}(t)$ is the intensity of the final channel as a function of time, and $I_{init_0}$ and $I_{final_0}$ are the intensities of the initial and final channels at t = 0.
* $Q = \dfrac{α}{1 - α}$ using the equation for α above

**Requirements:**
* Identify an *initial channel* (red, green, or blue) corresponding to the dominant color at room temperature/before reaction or transformation and a *final channel* corresponding to the dominant color at elevated temperatures/after reaction or transformation
* Store the filepath for an initial image showing the crystal before the reaction or transformation
* Create a list of temperatures as strings for graph titles at each temperature

```python
# Test load_init_image with demo data
temps = ['75', '90', '105', '120', '135']
filepath = 'demo_data/batch_images/TOPA-1-075/TOPA-1-075_b0v0t0z0c0-32x0-512y0-512.png' # enter filepath for baseline image .png file 

# Call the functions from crystal_tools for use in this code
from crystal_tools.reaction_analysis import load_init_image, calculate_rxn_progress

init_0, final_0 = load_init_image(filepath, 
                                  init_channel='blue', 
                                  final_channel='green') # Change initial and final channels if needed

# Loop through the different experiments in the dataframe
for i, (temp, df) in enumerate(zip(temps, df_list)):
    exp_name = f'Experiment {i+1} ({temp} °C)' # Format plot titles
    calculate_rxn_progress(df, init_0, final_0, exp_name, # Use the init_0 and final_0 values from the initial image
                           init_channel='blue', 
                           final_channel='green') # Change initial and final channels if needed
```

These functions output the initial intensities of the initial and final channel and subsequent plots of α versus time for each of the experiments. Scroll through, checking the y axes to make sure there are no major errors in normalization. Common errors can include drastically negative (< -0.1) values of α due to noise within the baseline image data. In this case, try using the second or third image as the baseline image rather than the first.

## Step 3: Detect Equilibration

This set of functions work to calculate an equilibrium

1. **detect_equilibration:** Use the approximate first derivative of temperature with respect to time to calculate the point at which the heating phase stops and the oscillation/equilibration phase begins
2. **analyze_temperature_series:** Fit a sinusoidal model to the oscillation/equilibration phase as determined using **detect_equilibration**
3. **calc_equilibrium:** Calculate an equilibrated temperature using either an exponential decay model or the sinusoidal model fitted to the equilibration stage in **analyze_temperature_series** and an equilibrium constant using an exponential decay model fitted to the reaction quotient Q

```python
# Call the functions from crystal_tools for use in this code
from crystal_tools.temperature_fitting import detect_equilibration, sinusoidal_model, analyze_temperature_series, exp_decay, calc_equilibrium

# Calculate equilibrium temperatures and constants for each dataframe in df_list
eq_temps, eq_consts = calc_equilibrium(df_list, 'exponential') # 'exponential' can be changed to 'sinusoidal' (default)
```

## Step 4: Create van't Hoff Plot

This function uses the list of equilibrated temperatures and equilibrium constants from Step 3 to create a van't Hoff plot. A regression is fitted to the van't Hoff plot using either the linear van't Hoff equation (which assumes no change in heat capapcity) or a 3-variable truncation of the extended van't Hoff equation from Clarke & Glew 1966 (which accounts for a change in heat capacity over the course of the reaction/transformation).
